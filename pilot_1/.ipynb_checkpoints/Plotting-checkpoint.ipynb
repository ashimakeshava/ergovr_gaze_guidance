```python
import json
import pandas as pd
from pandas.io.json import json_normalize
from pandas.api.types import CategoricalDtype
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
import pickle
import seaborn as sns
import sys
import time
from sklearn.preprocessing import LabelEncoder
```

```python
#DEFAULTS
PLOT_DIR = './PLOTS_extendedEdges'
COLORS = {'g': '#CAE6CB', 'r': '#E6CACA', 'b': '#C9D4F5'}
sns.set(context = "talk", style="whitegrid", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
os.makedirs(os.path.dirname('./PLOTS_extendedEdges/'), exist_ok=True)
```

```python
#HELPERS
def pickleData(filepath,filename, data):
    os.makedirs(os.path.dirname(filepath), exist_ok=True)
    with open(str(filepath+filename),'wb') as fh:
        pickle.dump(data,fh)
        
def read_dataFrameFromFile(filename):
    with open(filename, 'rb') as fh:
        dataframe = pickle.load(fh)
    return dataframe
```

# Read ROI labeled data

```python
# frames = read_dataFrameFromFile('./cleaned_data/frames_labeled_final')
frames = read_dataFrameFromFile('./cleaned_data/frames_labeled_withExtendedEdge')
# frames['hitCubeType'] = frames['hitCubeType'].map({0:'Reference Cube', 1:'Transfer Cube'})
frames['trialID'] = 24*(frames['blockID'] - 1) + frames['trialID']
frames.columns
```

```python
frames.groupby(['subjectID','trialID','hitCubeType'])['hitLoc'].count().reset_index().plot()
```

```python
frames.drop(columns= ['diffCurrentCubeToOriginal', 'endAlignment', 'startGrabbing', 'hit.cubeID', 'modelName',\
                     'blockID', 'hitPos', 'eyePos', 'gazeDir','transCubePos',\
                      'refCubePos',  'hitLoc'], inplace = True)
```

```python
# frames['hitCubeType'] = frames['hitCubeType'].map({0:'Reference', 1: 'Transfer'})
q_idx = frames.query('refCubeSz == "S" | transCubeSz == "S"').index
frames.drop(q_idx, inplace = True)
cat_vars = ['transCubeSz','refCubeSz', 'ROI_Label','hitCubeType',
            'hitCubeSz', 'hitLoc','alignmentType', 'isMoving', 'subjectID' ]
frames_enc = frames.copy()
frames_enc.drop(columns=['endAlignment', 'startGrabbing', 'hit.cubeID', 'modelName', 'trialID',
      'blockID', 'hitPos', 'eyePos', 'gazeDir', 'transCubePos', 'ROI_Hit'], inplace=True)
frames_enc.dropna(subset=['hitCubeType'], inplace = True)
frames_enc.fillna('NaN', inplace = True)

```

```python
for l in cat_vars:
    frames_enc[l] = LabelEncoder().fit_transform(frames_enc[l])
```

```python
frames_enc
```

```python
sns.set(context = "talk", style="whitegrid", palette="Accent", font_scale=1, rc={'figure.figsize':(16,12)})
sns.heatmap(frames_enc.corr(method = 'spearman').round(decimals=2), linewidths=0.1,square=True,linecolor='white',\
            annot=True, center=0, vmin = -1, vmax = 1)
plt.title('Correlations between variables')
plt.savefig(PLOT_DIR+'/CORR_HEATMAP.png', quality=90, dpi=1000)
```

```python
sns.pairplot(frames_enc,hue='alignmentType')
```

## START THE PLOTTING!

### FRAME RATE

```python
times = frames.reset_index().groupby(['subjectID','blockID'])['timeStamp']\
                                    .apply(lambda x: 1/np.mean((x - x.shift())/np.timedelta64(1,'s')))\
                                    .rename('Frame Rate')\
                                    .reset_index()
# times = times.groupby(level = 0).agg({'Frame Rate':['mean','std']}).reset_index()
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(14,8.27)})
sns.barplot(data=times, x='subjectID', y='Frame Rate', color=COLORS['g'],errwidth=2, capsize=0,\
                                                                                saturation=0.5,ci='sd')

plt.ylabel('Frame rate (Hz)')
plt.title('Frame Rate Over Subjects')
plt.savefig(PLOT_DIR+'/framerate_persubject.png', quality=90, dpi=1000)
```

### COLLIDER HITS PER ALIGNMENTTYPE

```python
hits = frames.groupby(['alignmentType','subjectID'])['hitCubeType']\
                                            .apply(lambda x: x.notnull().mean())\
                                            .reset_index(name= 'Hit Rate')
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.barplot(data = hits, x='alignmentType',y='Hit Rate', color=COLORS['b'],saturation=0.5,\
                                                                        errwidth=2, capsize=0,\
                                                                        ci='sd')
plt.ylim([0,1])
plt.ylabel('Cube Hit Rate')
plt.title('Hit Rate Over Subjects')
plt.savefig(PLOT_DIR+'/hitrate_percondition.png', quality=90, dpi=1000)
```

```python
frames['hitCubeType'] = frames['hitCubeType'].map({0: 'Reference Cube', 1: 'Transfer Cube'})
df = frames.copy()
# df['ROI_Label'] = df['ROI_Label'].fillna('None')
df['hitCubeType'] = df['hitCubeType'].fillna('None')
# df['hitCubeType'] = df['hitCubeType'].map({'None': 'None'})
hits = df.groupby([ 'subjectID','alignmentType','hitCubeType'])\
                                            .agg({'trialID': 'count'})


hits = hits.groupby(level=[0,1]).apply(lambda x: x/x.sum())\
                                            .rename(columns={'trialID':'% of Hits'})\
                                            .reset_index()

sns.set(context = "poster", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.barplot(data=hits, x='alignmentType',y='% of Hits', hue ='hitCubeType',saturation=0.5,\
                                                                        errwidth=2, capsize=0,\
                                                                        ci='sd',  dodge=True)
# # plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
# # plt.setp(ax.get_legend().get_title(), fontsize='15') # for legend title
plt.legend(title = 'Hit Type',loc='best')
plt.title('Rate of Raycast Hits')
plt.ylabel('Collider Hit Rate')
plt.savefig(PLOT_DIR+'/hitrate_oversubject_overconds.png', quality=90, dpi=1000)
```

```python

# frames['hitCubeType'] = frames['hitCubeType'].map({0: 'Reference Cube', 1: 'Transfer Cube'})
grp = frames.groupby(['subjectID','alignmentType','hitCubeType' ])\
                                                    .agg({'trialID': 'count'})
#                                                     .rename(columns={'trialID':'Number of Hits'})\
#                                                     .reset_index()
grp_pct = grp.groupby(level = [0,1]).apply(lambda x: x/x.sum())\
                                                    .rename(columns={'trialID':'% of Hits'})\
                                                    .reset_index()
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.violinplot(data=grp_pct, x='alignmentType',y='% of Hits', hue ='hitCubeType',saturation=0.5,\
                                                                        errwidth=3, capsize=0,\
                                                                        ci='sd',  split=True)
# # plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
# # plt.setp(ax.get_legend().get_title(), fontsize='15') # for legend title
plt.legend(title = 'Hit Cube Type',loc='lower right')
plt.title('Hits per Cube Type')
plt.ylabel('Hit Rate')
plt.savefig(PLOT_DIR+'/hitrate_oversubject_overconds_cubes.png', quality=90, dpi=1000)
```

```python
sns.violinplot(data=grp_pct, x='alignmentType',y='% of Hits', hue ='hitCubeType',\
               scale='count',inner='stick',split=True,dodge=True)
plt.legend(title = 'Hit Cube Type',loc='best')
plt.title('Hit Rate Over Subjects')
plt.ylabel('Hit Rate')
plt.savefig(PLOT_DIR+'/hitrate_oversubject_overconds_violin.png', quality=90, dpi=1000)
```

```python
grp = frames.groupby([ 'subjectID','alignmentType','hitCubeType', 'isMoving'])\
                                                    .agg({'trialID': 'count'})\
                                                    .rename(columns={'trialID':'Hit Rate'})\
                                                    .reset_index()
# grp = grp.groupby(level = [0,1, 3]).apply(lambda x: x/x.sum())\
#                                                     .rename(columns={'trialID':'Hit Rate'})\
#                                                     .reset_index()
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.catplot(x='alignmentType', y='Hit Rate', data=grp, hue ='hitCubeType', 
                col='isMoving', kind='violin',dodge=0.25, join = False, sharey=True, split=True, inner='stick')
# # plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
# # plt.setp(ax.get_legend().get_title(), fontsize='15') # for legend title
# plt.ylim([0, 1])
plt.subplots_adjust(top=0.8)
plt.suptitle("Proportion of Raycast Hits",fontsize=25)
plt.savefig(PLOT_DIR+'/hitrate_oversubject_overconds_isMoving.png', quality=90, dpi=1000)
```

```python
grp = frames.groupby([ 'subjectID','alignmentType','hitCubeType', 'hitCubeSz'])\
                                                    .agg({'trialID': 'count'})\
                                                    .rename(columns={'trialID':'Number of Hits'})\
                                                    .reset_index()

# grp_pct = grp.groupby(level = [0,1,2]).apply(lambda x: x/x.sum())\
#                                                     .rename(columns={'trialID':'Hit Rate'})\
#                                                     .reset_index()
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.catplot(x='alignmentType', y='Number of Hits', data=grp, hue ='hitCubeType', 
                col='hitCubeSz',kind='point',dodge=0.2, markers=["o", "x"], sharey=True)
# # plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
# # plt.setp(ax.get_legend().get_title(), fontsize='15') # for legend title
# plt.ylim([0, 1])

plt.subplots_adjust(top=0.8)
plt.suptitle("Raycast Hits Based on Cube Size",fontsize=25)
plt.savefig(PLOT_DIR + '/hitrate_oversubject_overconds_cubeSz.png', quality=90, dpi=1000)
```

```python
grp = frames.groupby([ 'subjectID','alignmentType','hitCubeType', 'ROI_Label']).agg({'trialID': 'count'})\
                                                    .rename(columns={'trialID':'Number of Hits'})\
                                                    .reset_index()

# grp = pd.melt(grp, id_vars=['subjectID', 'alignmentType', 'hitCubeType'], var_name='Hit_Location', value_name='Number of Hits')
# grp = grp.groupby(level = [0,1,4]).apply(lambda x: x/x.sum())\
#                                                     .rename(columns={'trialID':'Hit Rate'})\
#                                                     .reset_index()
sns.set(context = "talk", style="white", palette="Accent", font_scale=1, rc={'figure.figsize':(11.7,8.27)})
sns.catplot(x='alignmentType', y='Number of Hits', data=grp, col ='hitCubeType', 
               hue='ROI_Label' , kind='point',dodge=0.3, join = False, sharey=True)

plt.subplots_adjust(top=0.8)
plt.suptitle("Raycast Hits Based on Location on the Cubes",fontsize=25)
plt.savefig(PLOT_DIR + '/hitrate_oversubject_overconds_hitloc.png', quality=90, dpi=1000)
```
